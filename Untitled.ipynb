```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
manuale= pd.read_csv('manuale.csv' ,sep=';' , thousands="." )
training= pd.read_csv('training.csv' ,sep=';' , thousands="." )
```

```python
df=pd.DataFrame(manuale)
```

```python
df
```

```python
df.plot(kind="scatter",x="Partite Giocate",y="Minuti Giocati")
```

```python
df.plot(kind="box",y="Minuti Giocati")
```

```python
modello= LinearRegression()
X = df[['Età', 'Partite Giocate','Reti']]  
y = df['Minuti Giocati']    
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print("Coefficiente:", model.coef_)
print("Intercetta:", model.intercept_)
print("R2 score:", model.score(X_test, y_test))
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(y_test, y_pred, color='blue', alpha=0.6)
plt.plot([y_test.min(),y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.xlabel("Valori reali")
plt.ylabel("Valori previsti")
plt.title("Valori reali vs. previsti")
plt.grid(True)
plt.show()
```
